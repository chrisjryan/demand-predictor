```python
#!/usr/bin/env python

import datetime
import matplotlib.pyplot as plt
import math
import numpy
import pytz
import dataprep
import predict
import seaborn


def plot_raw(hours, usage, tz_str = 'UTC'):
    """
    Plot a time series of each hour-window in the data set along with its 
    corresponding timestamp count. 'hours' variable is a list of datetime 
    objects for each hour within some range (even those for which there were 
    0 timestamps binned). 'usage' is a corredponding count of timestamps 
    recorded within each hour window in the hours list. These 2 lists can be 
    acquired from dataprep.prepare() or dataprep.bin_timestamp().
    """
    
    numweeks = int(numpy.ceil((max(hours)-min(hours)).days/7.0))

    %matplotlib inline
    starthour = hours[0]
    datetime_axis = [starthour + datetime.timedelta(hours = x) for x in range(24*7*numweeks)]
    ymax = math.ceil(max(usage)/10.0)*10
    for w in range(numweeks):
        plt.subplot(numweeks,1,w+1)
        plt.xticks(rotation='vertical', size='large')
        plt.xlim(xmin=datetime_axis[w*24*7], xmax=datetime_axis[w*24*7 + 24*7-1])
        plt.ylim(0,ymax)
        plt.title('User logins, timezone: ' + tz_str )
        plt.xlabel('(date labels correspond to midnight)')
        plt.plot(hours, usage)

    fig = plt.gcf()
    fig.set_size_inches(12,10*9)
        
    
def plot_avg(hourly_usage_stats, tz_str = 'UTC'):
    """
    Plot the average timestamp count for each hour in a week. 
    'hourly_usage_stats' is a 2-dimensional list of namedtuples containing 
    mean and sample variance for each of these hours. They are indexed by 
    weekday and hour. This list is typically the output of 
    predict.average_all_hours()
    """
    
    hourly_usage_stats_flat = [(hour.mean, numpy.sqrt(hour.var)) for weekday in hourly_usage_stats for hour in weekday]
    mean, stderr = (numpy.array(t) for t in zip(*hourly_usage_stats_flat))
    
    %matplotlib inline
    plt.title('Average user logins, timezone: ' + tz_str )
    plt.plot(mean)
    plt.fill_between(range(len(mean)), mean-stderr, mean+stderr, color='#CC4F1B', alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
    plt.xticks(numpy.arange(7)*24, ('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'), rotation='vertical', size='large')
    plt.ylim(0)
    plt.xlabel('(date labels correspond to midnight)')

    fig = plt.gcf()
    fig.tight_layout()
    fig.set_size_inches(12,6)
    return fig
    
```

```python

if __name__ == "__main__":
    hours, usage = dataprep.prepare('hourly_demand_prediction_challenge.json')
    # sort by date & plot:
    hours, usage = zip(*sorted(zip(hours, usage), key = lambda hu:hu[0]))
    plot_raw(hours, usage)
    
```
